### Задание 1 - Создайте столбец source_type

```python
df = pd.read_csv('C:/Users/Lenovo/Downloads/visit_log.csv', sep = ';')
```

```python
df.head()
```

```python
def type_maker(row):
    if (row['traffic_source'] == 'yandex') | (row['traffic_source'] == 'gogle'):
        return 'organic'
    elif (row['traffic_source'] == 'paid') | (row['traffic_source'] == 'mail'):
        if row['region'] == 'Russia':
            return 'ad'
        else:
            return 'other'
    else:
        return row['traffic_source']
```

```python
df['source_type'] = df.apply(type_maker, axis=1)
```

### Задание 3 - Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок

```python
raitings = pd.read_csv('C:/Users/Lenovo/Downloads/ratings.csv')
```

```python
raitings['date'] = pd.to_datetime(raitings['timestamp'], unit = 's')
```

```python
raitings.head()
```

```python
raitings.groupby('userId') \
        .agg({'rating':'count','date':lambda column: column.max() - column.min()}) \
        .query('rating > 100') \
        .rename(columns = {'rating':'mark_count', 'date':'life_time'}) \
        .describe()
```
